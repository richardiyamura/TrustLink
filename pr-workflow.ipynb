````python
# PR Workflow: Create and Push a Detailed Pull Request
This notebook guides you through creating a feature branch, committing your changes, pushing them to GitHub, and opening a detailed pull request.
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 1. Setup Git Repository and Remote
Verify the repository status, configure your user information, and confirm that the remote origin is pointing to `https://github.com/Haroldwonder/TrustLink`.
</VSCode.Cell>

<VSCode.Cell language="python">
# Verify repository status and remote origin
import subprocess
from pathlib import Path

repo = Path('/workspaces/TrustLink')
print('Repository path:', repo)
print('Git status:')
print(subprocess.run(['git', 'status', '--short'], cwd=repo, capture_output=True, text=True).stdout)
print('Current branch:')
print(subprocess.run(['git', 'branch', '--show-current'], cwd=repo, capture_output=True, text=True).stdout)
print('Remote origin:')
print(subprocess.run(['git', 'remote', 'get-url', 'origin'], cwd=repo, capture_output=True, text=True).stdout)
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 2. Create a Feature Branch
Create and switch to a new branch for this work. Use a descriptive naming convention like `fix/pagination-consistency`.
</VSCode.Cell>

<VSCode.Cell language="python">
repo = Path('/workspaces/TrustLink')
branch_name = 'fix/pagination-consistency'
print('Creating new branch:', branch_name)
print(subprocess.run(['git', 'checkout', '-b', branch_name], cwd=repo, capture_output=True, text=True).stdout)
print('Branch created and switched to:')
print(subprocess.run(['git', 'branch', '--show-current'], cwd=repo, capture_output=True, text=True).stdout)
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 3. Stage and Commit Changes
Stage all modified files and create a descriptive commit message for the pagination fix.
</VSCode.Cell>

<VSCode.Cell language="python">
repo = Path('/workspaces/TrustLink')
files_to_stage = ['src/query.rs', 'src/lib.rs', 'src/test.rs', 'README.md']
print('Staging files:')
print('\n'.join(files_to_stage))
print(subprocess.run(['git', 'add'] + files_to_stage, cwd=repo, capture_output=True, text=True).stdout)
commit_message = 'fix(query): add cursor-based pagination for get_attestations_in_range and document deletion-safe workflow'
print('Commit message:')
print(commit_message)
print(subprocess.run(['git', 'commit', '-m', commit_message], cwd=repo, capture_output=True, text=True).stdout)
print('Git log (latest commit):')
print(subprocess.run(['git', 'log', '-1', '--oneline'], cwd=repo, capture_output=True, text=True).stdout)
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 4. Push Branch to Remote
Push the new feature branch to the remote repository and configure upstream tracking.
</VSCode.Cell>

<VSCode.Cell language="python">
repo = Path('/workspaces/TrustLink')
print(subprocess.run(['git', 'push', '--set-upstream', 'origin', branch_name], cwd=repo, capture_output=True, text=True).stdout)
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 5. Create a Detailed Pull Request
Use GitHub CLI or the GitHub web interface to open a PR from the feature branch into `main`.

Suggested title:
- `fix(pagination): resilient cursor-based pagination for get_attestations_in_range`

Suggested description:
- Implement cursor-based pagination via `get_attestations_in_range_after`.
- Preserve legacy `get_attestations_in_range` and document offset-based limitations.
- Add tests for cursor pagination and GDPR deletion resilience.
- Update README examples to demonstrate the new API.
</VSCode.Cell>

<VSCode.Cell language="python">
repo = Path('/workspaces/TrustLink')
pr_title = 'fix(pagination): resilient cursor-based pagination for get_attestations_in_range'
pr_body = '''
### Summary
This PR introduces cursor-based pagination for `get_attestations_in_range` results, ensuring deterministic traversal when GDPR deletions remove earlier attestation entries between paginated requests.

### Changed
- Add `get_attestations_in_range_after` contract API and query helper.
- Preserve legacy offset-based `get_attestations_in_range` for compatibility.
- Document cursor-pagination behavior and deletion-safe usage in `README.md`.
- Add regression tests covering deletion between pages and invalid cursor handling.

### Testing
- Added tests in `src/test.rs`.
- Existing pagination and range query tests remain intact.
'''
print('Prepare to create PR with GitHub CLI:')
print(f'gh pr create --fill --title "{pr_title}" --body "{pr_body}"')
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 6. Merge PR and Push to Main
Once the PR is approved, merge it into `main`, then pull or push the updated `main` branch.

If using GitHub CLI:
```bash
gh pr merge --merge
```

Then:
```bash
git checkout main
git pull origin main
git push origin main
```
</VSCode.Cell>
````
